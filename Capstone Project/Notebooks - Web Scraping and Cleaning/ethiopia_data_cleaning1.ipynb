```python
df_all_2.info()
```

```python
df_all_2['Destination Name'] = df_all_2['Destination Name'].replace(to_replace = 'Chebera-Churchura National Park', 
                                                                    value = 'Chebera Churchura National Park')
df_all_2.head()
```

```python
destination_list = list(df_all_2['Destination Name'].unique())
for i in range(len(destination_list)):
    print(i + 1, destination_list[i])
```

```python
landmarks = ['Rock-Hewn Churches of Lalibela', 'Church of St. George', 'Abuna Yemata Guh church', 'Yemrehanna Kristos',
             'Debre Berhan Selassie Church', 'Azwa Mariam Monastery', 'Fasiladas Bath Timket', 'The Ruins of Aksum',
             'Holy Trinity Cathedral', 'Simien View', 'Asheton Maryam Monastery', 'St. George\'s Cathedral',
             'Northern Stelae Field', 'Monastery of Na’akuto La’ab', 'Entoto Maryam Church', 'Tomb of Adam',
             'Monastery of Debre Mariam', 'Church of Our Lady Mary of Zion', 'Tombs of Kings Kaleb and Gebre Meskal', 
             'The Mausoleum of Menelik II', 'Dasenech Village', 'King Ezana\'s inscription', 'St Mary of Zion',
             'Harar Jegol Wall', 'Adadi Maryam', 'Monastery of Debre Damo', 'Konso Cultural Centre', 
             'Bilbala St. George Rock Hewn Church', 'Unity Park', 'Medhane Alem Cathedral', ] 
nature_parks = ['Simien Mountains National Park', 'Omo National Park and River', 'Gambella National Park', 'Gheralta Rock',
                'Abune Yosef', 'Nechisar National Park', 'Abidjatta-Shalla National Park', 'Bale Mountains',
                'Menagesha Suba Forest Park', 'Yabelo Wildlife Sanctuary', 'Gibe Sheleko National Park', 'BlueSapphire Tours',
                'Yerer Mountain', 'Bale Mountains National Park', 'Senkele Wildlife Sanctuary',
                'Hippo watching on Lake Awassa', 'Ras Dashen', 'Mount Bwahit', 'Mago National Park', 
                'Chebera Churchura National Park', 'Mesganaw', 'Ezana Park', 'Entoto Natural Park', 
                'Senkele Swayne\'s Hartebeest Sanctuary']
museums = ['National Museum of Ethiopia', 'Red Terror Martyrs Memorial Museum', 'Ethnological Museum', 'Harar Museum',
           'Emperor Yohannes IV Palace', 'Berhan Ethiopa Cultural Center Plc', 'Casa Museo di Rimbaud', 
           'Church of Debre Sina Maryam', 'Asni Gallery', 'Addis Fine Art Gallery', 'Wakanda Art Gallery',
           'Museum of Modern Art/Goethe Institut Ethiopia', 'Addis Ababa Museum', 'Archeological Museum', 
           'National Theatre Art Gallery', 'South Omo Research Centre Museum', 'Zoma Museum', 'Lions Art Gallery',
           'Lalibela Gemstones and Jewellery', 'The Fine Art Gallery', 'Wukro Museum', 'Lela Gallery']
```

```python
len(landmarks) + len(nature_parks) + len(museums)
```

```python
def destinationType(row):
    if row['Destination Name'] in landmarks:
        return 'Landmarks'
    elif row['Destination Name'] in nature_parks:
        return 'Nature/Parks'
    elif row['Destination Name'] in museums:
        return 'Museums'
    
df_all_2['Destination Type'] = df_all_2.apply(destinationType, axis = 1)
df_all_2.head()
```
